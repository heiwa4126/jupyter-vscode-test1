## 活性化関数 (Activation function)

### シグモイド関数
$$\sigma(x) = \frac{1}{1 + e^{-x}}$$

-∞ ~ +∞ の数値を、0.0~1.0 の範囲の数値に変換する関数

### 双曲線正接関数

$$\tanh(x) = \frac{e^{x} - e^{-x}}{e^{x} + e^{-x}}$$

-∞ ~ +∞ の数値を、-1.0~1.0 の範囲の数値に変換する関数

### ReLU (Rectified Linear Unit)

$$\text{ReLU}(x) = \max(0, x)$$

```python
x = np.arange(-8, 8, 0.1) # xの値(-8~8で0.1刻みで配列生成)
y1 = 1 / (1 + np.exp(-x) ) #  Sigmoid function (シグモイド関数)
# y2 = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x)) # tanh関数 (双曲線正接関数。タンエイチ、タンチ、ハイパボリックタンジェント)

# グラフの設定
plt.plot(x, y1, label="sigmoid")
plt.plot(x, np.tanh(x), label="tanh")
plt.plot(x, np.maximum(0, x), label="ReLU")
plt.xlim(-8, 8)  # x軸の範囲
plt.ylim(-1.25, 1.25) # y軸の範囲
plt.grid() # グリッド描画
plt.legend()  # 凡例を表示
plt.show() # グラフを出力
```
